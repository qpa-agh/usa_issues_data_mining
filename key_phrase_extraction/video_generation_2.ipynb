```python
for keyphrase in predefined_keyphrases:
    df[keyphrase] = df['full_text'].apply(
    lambda x: 1 if keyphrase in x else 0 
)
```

```python
df
```

```python
prepared_df = df.drop(columns=['representative_party', 'full_text']).groupby(['year', 'month']).sum().reset_index()
prepared_df['time'] = pd.to_datetime(prepared_df[['year', 'month']].assign(day=15))
prepared_df = prepared_df.drop(columns=['year', 'month'])
prepared_df.set_index('time', inplace=True)
prepared_df
```

```python
chosen_keypharses = set(prepared_df.max().sort_values(ascending=False)[:25].keys())
chosen_keypharses = set(prepared_df.idxmax().keys())
prepared_df.drop(columns=[keyphrase for keyphrase in predefined_keyphrases if keyphrase not in chosen_keypharses], inplace=True)
```
