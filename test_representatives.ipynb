```python
import pandas as pd
from representatives.evo import evo_extract_issues
from representatives.recordsContainer import recordsContainer_extract_issues
from utils.scraping_utils import get_html, parrarelize_processes
from wrappers import extract_issues_wrapper_proofed
```

```python
rep_df = pd.read_csv('data\cleaned_representatives.csv')[['state',	'district',	'name',	'party',	'committee',	'page_link']]
rep_df
```

```python
jobs = []
additional_info = []
for name, state, party, committee, link in zip(rep_df['name'], rep_df['state'], rep_df['party'], rep_df['committee'], rep_df['page_link']):
   
    if link[-1] != '/':
        link += '/'
    # print(link)
    jobs.append([link, evo_extract_issues])
    additional_info.append([name, state, party, committee, link]) 

articles = [] 
for id, result in parrarelize_processes(extract_issues_wrapper_proofed, jobs, n_executors=10):
    record = {}
    record['representative_name'] = additional_info[id][0]
    record['representative_state'] = additional_info[id][1]
    record['representative_party'] = additional_info[id][2]
    record['representative_committee'] = additional_info[id][3]
    record['link'] = additional_info[id][4]
    record['issues_catched'] = None
    if len(result) != 0:
        record['issues_catched'] = [issue['name'] for issue in result]
    print(record)
    articles.append(record)

evo_df = pd.DataFrame.from_dict(articles)
evo_df

evo_df.to_csv('data/evo_representatives.csv')
evo_df
```
